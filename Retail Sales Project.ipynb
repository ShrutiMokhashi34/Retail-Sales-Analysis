```python
#Rename Column Names
retail.columns
retail.columns = retail.columns.str.lower()
retail.columns = retail.columns.str.replace(' ', '_')
retail.columns
```

```python
retail.head()
```

```python
#New Columns to measure Discount, Selling Price, Profit
retail['discount'] = retail['list_price']*retail['discount_percent']*0.01
retail['sell_price'] = retail['list_price'] - retail['discount']
retail['profit'] = retail['sell_price'] - retail['cost_price']
```

```python
retail.head()
```

```python
#retail.dtypes
#Converting order_date from object to datetime 
retail['order_date'] = pd.to_datetime(retail['order_date'], format='%Y-%m-%d')
retail.dtypes
```

```python
#Removing Unwanted Columns
retail.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace=True) #inplace -> Make the changes in the original dataset
```
